# Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data Collection and Preprocessing

```python
data = pd.read_csv('creditcard.csv')
data
```

```python
# dataset information
data.info()
```

```python
# drop last row
data = data.drop(73376)
```

```python
# check for missing values
data.isnull().sum()
```

```python
# distribution of target column
data['Class'].value_counts()
```

```python
# seperating the data for analysis
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
data.groupby('Class').mean()
```

Under-sampling

```python
legit_sample = legit.sample(n=len(fraud))
print(legit_sample.shape, fraud.shape)
```

```python
# concating two dataframes
new_data = pd.concat([legit_sample, fraud], axis=0)
new_data
```

```python
# target value counts
new_data['Class'].value_counts()
```

```python
# get mean
new_data.groupby('Class').mean()
```

```python
# splitting data into target and features
X = new_data.drop(columns='Class', axis=1)
Y = new_data['Class']
```

```python
print(X)
print(Y)
```

```python
# splitting into train and test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model Training

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
# accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Training data accuracy is: ', training_data_accuracy)
```

```python
# accuracy on testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Testing data accuracy is: ', testing_data_accuracy)
```
