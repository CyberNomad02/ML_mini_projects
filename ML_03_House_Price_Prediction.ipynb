# Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing Boston House Price Dataset

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
boston = pd.read_csv('/content/boston_housing.csv', header=None, delimiter=r"\s+", names=column_names)
print(boston)
```

```python
boston.isnull().sum()
```

```python
boston.describe()
```

Correlation

```python
correlation = boston.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(boston.corr().abs(),  annot=True)
```

Splitting the target and labels

```python
X = boston.drop(['MEDV'], axis=1)
Y = boston['MEDV']

print(X)
print(Y)
```

# Splitting into training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model Training

```python
model = XGBRegressor()
```

```python
model.fit(X_train, Y_train)
```

# Model Evaluation

```python
# Prediction on training data
training_data_prediction = model.predict(X_train)
score1 = metrics.r2_score(Y_train, training_data_prediction)
score2 = metrics.mean_absolute_error(Y_train, training_data_prediction)
print('R Squared value = ', score1)
print('Mean Absolute Error = ', score2)
```

```python
# Prediction on testing data
testing_data_prediction = model.predict(X_test)
score1 = metrics.r2_score(Y_test, testing_data_prediction)
score2 = metrics.mean_absolute_error(Y_test, testing_data_prediction)
print('R Squared value = ', score1)
print('Mean Absolute Error = ', score2)
```

# Visualisation of Actual and Predicted Prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```
